```python
# feature engineering
def get_lat_long(location):
    from geopy.geocoders import Nominatim
    if ',' in str(location):
        location = location.split(',')[1]
    try:
        geolocator = Nominatim(user_agent='GA Project 4')
        location_p = geolocator.geocode(location)
        print(location)
        print((location_p.latitude, location_p.longitude))
        return location_p.latitude, location_p.longitude
    except:
        return None, None

lat_long = df.job_location.map(get_lat_long)
```

```python
df['Latitude'] = pd.DataFrame(lat_long).apply(lambda x: x[0][0], axis=1)
df['Longitude'] = pd.DataFrame(lat_long).apply(lambda x: x[0][1], axis=1)
```

```python
df.job_employment_type.value_counts()
```

```python
df['Job_Permanent'] = df.job_employment_type.map(lambda x:1 if 'permanent' in x.lower() else 0)
df['Job_FullTime'] = df.job_employment_type.map(lambda x:1 if 'full time' in x.lower() else 0)
df['Job_Contract'] = df.job_employment_type.map(lambda x:1 if 'contract' in x.lower() else 0)
df['Job_Temporary'] = df.job_employment_type.map(lambda x:1 if 'temporary' in x.lower() else 0)
df['Job_PartTime'] = df.job_employment_type.map(lambda x:1 if 'part time' in x.lower() else 0)
df['Job_Internship'] = df.job_employment_type.map(lambda x:1 if 'internship' in x.lower() else 0)

df.head()
```

```python
import numpy as np
df_subset_predict_salary = df[~df.min_salary.isnull()]
df_subset_predict_salary = df_subset_predict_salary[df_subset_predict_salary.Latitude != np.NaN]
target = df_subset_predict_salary.min_salary
predictor = df_subset_predict_salary[[x for x in df_subset_predict_salary.columns if 'Job' in x]]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold

kf = KFold(5)
scores = []
for train_index, test_index in kf.split(predictor, target):
    X_train = predictor.iloc[train_index]
    y_train = target.iloc[train_index]
    X_test = predictor.iloc[test_index]
    y_test = target.iloc[test_index]
    model_lcv = LinearRegression()
    model_lcv.fit(X_train, y_train)
    scores.append(model_lcv.score(X_test, y_test))
np.mean(scores)
```
